# MLB 球員下一年薪資預測專案

# 專案目標與商業意義
- 目標：用 t 年表現預測 t+1 年薪資（log 轉換）
- 意義：幫助球團評估合約價值、仲裁預測、自由市場參考

# 基礎環境設定

```python
import pandas as pd
import numpy as np
from pybaseball import batting_stats, pitching_stats, lahman, fielding_stats # 新增守位資料
import os
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']


# 姓名清洗函數 方便合併
def clean_name(name): 
    if isinstance(name, str):
        return name.lower().replace(" ", "").replace(".", "").replace("'", "")
    return ""
```

# 原始數據抓取

```python
# 抓取打者數據
B_RAW_DATA_FILE = 'data/mlb_raw_batting_2013_2023.csv'

if not os.path.exists(B_RAW_DATA_FILE):
    print("正在抓取打者原始數據 (2013-2023)...")
    b_raw_stats = batting_stats(2013, 2023, split_seasons=True)
    b_raw_stats.to_csv(B_RAW_DATA_FILE, index=False)
    print(f"原始數據已儲存至 {B_RAW_DATA_FILE}")
else:
    b_raw_stats = pd.read_csv(B_RAW_DATA_FILE)
    print("已從本地讀取打者原始數據。")

b_raw_stats['name_clean'] = b_raw_stats['Name'].apply(clean_name)

# 抓取投手數據
P_RAW_DATA_FILE = 'data/mlb_raw_pitching_2013_2023.csv'

if not os.path.exists(P_RAW_DATA_FILE):
    print("正在抓取投手原始數據 (2013-2023)...")
    p_raw_stats = pitching_stats(2013, 2023, split_seasons=True, qual = 0)
    p_raw_stats.to_csv(P_RAW_DATA_FILE, index=False)
    print(f"原始數據已儲存至 {P_RAW_DATA_FILE}")
else:
    p_raw_stats = pd.read_csv(P_RAW_DATA_FILE)
    print("已從本地讀取投手原始數據。")

p_raw_stats['name_clean'] = p_raw_stats['Name'].apply(clean_name)

# 新增：抓取野手守備數據
F_RAW_DATA_FILE = 'data/mlb_raw_fielding_2013_2023.csv'

if not os.path.exists(F_RAW_DATA_FILE):
    print("正在抓取野手守備原始數據 (2013-2023)...")
    f_raw_stats = fielding_stats(2013, 2023, split_seasons=True)
    f_raw_stats.to_csv(F_RAW_DATA_FILE, index=False)
    print(f"原始數據已儲存至 {F_RAW_DATA_FILE}")
else:
    f_raw_stats = pd.read_csv(F_RAW_DATA_FILE)
    print("已從本地讀取野手守備原始數據。")

f_raw_stats['name_clean'] = f_raw_stats['Name'].apply(clean_name)


# 讀取薪資數據 (https://www.kaggle.com/datasets/christophertreasure/mlb-player-salaries-2011-2024)
salary_df = pd.read_csv('data/mlb_salary_data.csv', encoding='latin1')
print("已從本地讀取薪資原始數據。")

salary_df ['name_clean'] = salary_df ['Name'].apply(clean_name)

# 讀取年資數據 (https://www.kaggle.com/datasets/dalyas/lahman-baseball-database/data)
people_path = 'data/People.csv'
people = pd.read_csv(people_path)
print("已從本地讀取年資原始數據。")

people['name_clean'] = (people['nameFirst'] + people['nameLast']).apply(clean_name)
```

# 資料合併

```python
# 處理薪資表：產生「下一年薪資」視角
salary_df = salary_df.sort_values(['name_clean', 'Year'])

# 處理薪資符號
salary_df['Salary'] = salary_df['Salary'].replace({r'[\$,]': ''}, regex=True).astype(float)

# 在薪資表中產生下一年薪資（用來合併）
salary_df['next_year_salary'] = salary_df.groupby('name_clean')['Salary'].shift(-1)
salary_df['next_year_log_salary'] = np.log1p(salary_df['next_year_salary'])

# 計算 debut year
people['debut'] = pd.to_datetime(people['debut'], errors='coerce')
people['debut_year'] = people['debut'].dt.year.fillna(0).astype(int)

# 合併薪資 (left join，用 name_clean + Year)
batting_df = b_raw_stats.merge(
    salary_df[['name_clean', 'Year', 'Team', 'Salary', 'Name', 'next_year_salary', 'next_year_log_salary']],
    left_on=['name_clean', 'Season'],
    right_on=['name_clean', 'Year'],
    how='left',
    suffixes=('', '_sal')
)

pitching_df = p_raw_stats.merge(
    salary_df[['name_clean', 'Year', 'Team', 'Salary', 'Name', 'next_year_salary', 'next_year_log_salary']],
    left_on=['name_clean', 'Season'],
    right_on=['name_clean', 'Year'],
    how='left',
    suffixes=('', '_sal')
)

# 合併 debut_year
people_subset = people[['name_clean', 'debut_year']].drop_duplicates()
batting_df = batting_df.merge(
    people_subset,
    on='name_clean',
    how='left'
)

pitching_df = pitching_df.merge(
    people_subset,
    on='name_clean',
    how='left'
)

# 計算「當季結束時的服務年資」（這是用來決定隔年薪資類別的關鍵）
batting_df['service_time'] = (batting_df['Year'] - batting_df['debut_year']).clip(lower=0)
pitching_df['service_time'] = (pitching_df['Year'] - pitching_df['debut_year']).clip(lower=0)

# 分類隔年適用的年資類別（pre-arb / arb / FA）
def classify_for_next_year(service_end_current):
    if np.isnan(service_end_current):
        return 'unknown'
    if service_end_current < 3:
        return 'pre-arb'
    elif service_end_current < 6:
        return 'arb'
    else:
        return 'FA'

batting_df['status_next_year'] = batting_df['service_time'].apply(classify_for_next_year)
pitching_df['status_next_year'] = pitching_df['service_time'].apply(classify_for_next_year)

# 做 one-hot encoding
batting_df = pd.get_dummies(
    batting_df,
    columns=['status_next_year'],
    prefix='status',          # 產生的欄位會變成 status_pre-arb, status_arb, status_FA
    drop_first=False          
)

pitching_df = pd.get_dummies(
    pitching_df,
    columns=['status_next_year'],
    prefix='status',          # 產生的欄位會變成 status_pre-arb, status_arb, status_FA
    drop_first=False          
)

# 加入 lag_1_log_salary
batting_df['lag_1_log_salary'] = batting_df.groupby('name_clean')['Salary'].shift(1).pipe(np.log1p)
batting_df['lag_1_log_salary'] = batting_df['lag_1_log_salary'].fillna(0)



# ===================================
# 新增：守位合併
# 只取第一個位置作為主要守位
fielding_df = f_raw_stats
fielding_df['primary_pos'] = fielding_df['Pos'].str.split('/').str[0].str.strip()

# 產生 one-hot
pos_dummies = pd.get_dummies(fielding_df['primary_pos'], prefix='pos')
fielding_df = pd.concat([fielding_df, pos_dummies], axis=1)

# 合併需要的欄位（守位）
fielding_merge = fielding_df[['name_clean', 'Season', 'primary_pos'] + list(pos_dummies.columns)].rename(columns={'Season': 'Year'})

# 合併（left join，保留打者所有資料，守位缺失填 'Unknown' 或 NaN）
batting_df = batting_df.merge(
    fielding_merge,
    on=['name_clean', 'Year'],
    how='left'
)

# 處理缺失值（無守備記錄的球員）
batting_df['primary_pos'] = batting_df['primary_pos'].fillna('Unknown')
batting_df[list(pos_dummies.columns)] = batting_df[list(pos_dummies.columns)].fillna(0)

print("合併後守位缺失比例：", batting_df['primary_pos'].isna().mean())
print(batting_df['primary_pos'].value_counts())





print(batting_df.head())
print(pitching_df.head())
print(batting_df[batting_df.name_clean=='mookiebetts'].sort_values('Year')[['name_clean', 'Year', 'Salary', 'next_year_salary', 'WAR', 'L-WAR']])
```

# 訓練模型

## 打者模型實驗記錄

### batter v1 - Baseline XGBoost
- 參數：n_estimators=300, lr=0.03, max_depth=5
- 訓練 R²: 0.9137
- 測試 R²: 0.2871
- 問題：嚴重過擬合

### batter v2 - 加強正則化
- 參數：n_estimators=600, lr=0.008, max_depth=4, reg_lambda=3
- 訓練 R²: 0.7186
- 測試 R²: 0.2933
- 改善：差距從 0.626 → 0.425

### batter v3 - 繼續加強
- 參數：n_estimators=800, lr=0.005, max_depth=3, reg_lambda=15.0
- 訓練 R²: 0.5858
- 測試 R²: 0.2820
- 改善：過擬合繼續被壓制（差距從 0.425 → 0.304）但測試 R² 不升反降（0.2933 → 0.2820）underfitting 開始出現

### batter v4 - 參數折衷版
- n_estimators=700, learning_rate=0.007, max_depth=4
- reg_lambda=12.0, reg_alpha=0.5, min_child_weight=6
- 訓練 R²: 0.6826 / RMSE: 0.7189
- 測試 R²: 0.2943 / RMSE: 1.0684
- 結論：目前最佳平衡點，測試表現最高

### batter v5 - 加入 lag_1_log_salary
- 新增特徵：lag_1_log_salary（前一年薪資 log，填 0 處理新人）
- 參數：延續 v4 折衷版
- 訓練 R²: 0.7100 / RMSE: 0.6872
- 測試 R²: 0.2568 / RMSE: 1.0965
- 結論：測試表現下降，可能因 lag_1 與 service_time/status 高共線，造成過度依賴慣性
- 下一步：移除或降權 lag_1，或先移除部分年資 one-hot 再試

### batter v6 - 移除 lag_1_log_salary
- 基於 v4 參數
- 訓練 R²: 0.6826 / RMSE: 0.7189
- 測試 R²: 0.2943 / RMSE: 1.0684
- 結論：與 v4 完全相同，證明 lag_1_log_salary 在此階段帶來負向影響（可能與年資共線或過度依賴慣性）
- 決定：暫不使用 lag_1_log_salary，鎖定 v4 作為基準

### batter v7 - 精簡到 13 個特徵
- 特徵：刪除 SB、SO、BB等
- 參數：延續 v4 折衷版
- 訓練 R²: 0.6680 / RMSE: 0.7352
- 測試 R²: 0.2974 / RMSE: 1.0660
- 結論：測試表現微升 0.0031，過擬合差距縮小至 0.3706，目前最佳版本

### batter v8 - 加入 EarlyStopping (callbacks)
- 特徵：13 個（v7 精簡版）
- 參數：延續 v4 折衷版 + EarlyStopping(rounds=60)
- 訓練 R²: 0.5600 / RMSE: 0.8465
- 測試 R²: 0.3082 / RMSE: 1.0579
- 結論：測試表現最高（0.3082），過擬合差距最小（0.2518），EarlyStopping 成功發揮作用

### batter v9 - 加入 Year 特徵（14 特徵）
- 特徵：13 個 + Year（數值特徵，捕捉時間趨勢）
- 參數：延續 v4 基準 + EarlyStopping (rounds=60)
- 訓練 R²: 0.6928 / RMSE: 0.7072
- 測試 R²: 0.3623 / RMSE: 1.0156
- 差距：0.3305
- 結論：加入 Year 後測試 R² 大幅提升 0.0541（從 0.3082 → 0.3623），時間趨勢對打者薪資有顯著貢獻

### batter v10 - 加入守位 one-hot (22 特徵)
- 特徵數：22（原 13 + Year + 8 個 pos_ one-hot）
- 參數：延續 v4 基準 + EarlyStopping (rounds=60)
- 訓練 R²: 0.6972 / RMSE: 0.7022
- 測試 R²: 0.3695 / RMSE: 1.0099
- 差距：0.3277
- 結論：加入守位後測試 R² 再升 0.0072（從 0.3623 → 0.3695）
- 決定：鎖定 v10 作為打者最終模型
