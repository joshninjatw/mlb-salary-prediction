# MLB 球員下一年薪資預測專案

# 專案目標與商業意義
- 目標：用 t 年表現預測 t+1 年薪資（log 轉換）
- 意義：幫助球團評估合約價值、仲裁預測、自由市場參考

# 基礎環境設定

```python
import pandas as pd
import numpy as np
from pybaseball import batting_stats, pitching_stats, lahman
import os
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']


# 姓名清洗函數 方便合併
def clean_name(name): 
    if isinstance(name, str):
        return name.lower().replace(" ", "").replace(".", "").replace("'", "")
    return ""
```

# 原始數據抓取

```python
# 抓取打者數據
B_RAW_DATA_FILE = 'data/mlb_raw_batting_2013_2023.csv'

if not os.path.exists(B_RAW_DATA_FILE):
    print("正在抓取打者原始數據 (2013-2023)...")
    b_raw_stats = batting_stats(2013, 2023, split_seasons=True)
    b_raw_stats.to_csv(B_RAW_DATA_FILE, index=False)
    print(f"原始數據已儲存至 {B_RAW_DATA_FILE}")
else:
    b_raw_stats = pd.read_csv(B_RAW_DATA_FILE)
    print("已從本地讀取打者原始數據。")

b_raw_stats['name_clean'] = b_raw_stats['Name'].apply(clean_name)

# 抓取投手數據
P_RAW_DATA_FILE = 'data/mlb_raw_pitching_2013_2023.csv'

if not os.path.exists(P_RAW_DATA_FILE):
    print("正在抓取投手原始數據 (2013-2023)...")
    p_raw_stats = pitching_stats(2013, 2023, split_seasons=True)
    p_raw_stats.to_csv(P_RAW_DATA_FILE, index=False)
    print(f"原始數據已儲存至 {P_RAW_DATA_FILE}")
else:
    p_raw_stats = pd.read_csv(P_RAW_DATA_FILE)
    print("已從本地讀取投手原始數據。")

p_raw_stats['name_clean'] = p_raw_stats['Name'].apply(clean_name)


# 讀取薪資數據 (https://www.kaggle.com/datasets/christophertreasure/mlb-player-salaries-2011-2024)
salary_df = pd.read_csv('data/mlb_salary_data.csv', encoding='latin1')
print("已從本地讀取薪資原始數據。")

salary_df ['name_clean'] = salary_df ['Name'].apply(clean_name)

# 讀取年資數據 (https://www.kaggle.com/datasets/dalyas/lahman-baseball-database/data)
people_path = 'data/People.csv'
people = pd.read_csv(people_path)
print("已從本地讀取年資原始數據。")

people['name_clean'] = (people['nameFirst'] + people['nameLast']).apply(clean_name)
```

# 資料合併

```python
# 處理薪資表：產生「下一年薪資」視角
salary_df = salary_df.sort_values(['name_clean', 'Year'])

# 處理薪資符號
salary_df['Salary'] = salary_df['Salary'].replace({r'[\$,]': ''}, regex=True).astype(float)

# 在薪資表中產生下一年薪資（用來合併）
salary_df['next_year_salary'] = salary_df.groupby('name_clean')['Salary'].shift(-1)
salary_df['next_year_log_salary'] = np.log1p(salary_df['next_year_salary'])

# 計算 debut year
people['debut'] = pd.to_datetime(people['debut'], errors='coerce')
people['debut_year'] = people['debut'].dt.year.fillna(0).astype(int)

# 合併薪資 (left join，用 name_clean + Year)
batting_df = b_raw_stats.merge(
    salary_df[['name_clean', 'Year', 'Team', 'Salary', 'Name', 'next_year_salary', 'next_year_log_salary']],
    left_on=['name_clean', 'Season'],
    right_on=['name_clean', 'Year'],
    how='left',
    suffixes=('', '_sal')
)

pitching_df = p_raw_stats.merge(
    salary_df[['name_clean', 'Year', 'Team', 'Salary', 'Name', 'next_year_salary', 'next_year_log_salary']],
    left_on=['name_clean', 'Season'],
    right_on=['name_clean', 'Year'],
    how='left',
    suffixes=('', '_sal')
)

# 合併 debut_year
people_subset = people[['name_clean', 'debut_year']].drop_duplicates()
batting_df = batting_df.merge(
    people_subset,
    on='name_clean',
    how='left'
)

pitching_df = pitching_df.merge(
    people_subset,
    on='name_clean',
    how='left'
)

# 計算「當季結束時的服務年資」（這是用來決定隔年薪資類別的關鍵）
batting_df['service_time'] = (batting_df['Year'] - batting_df['debut_year']).clip(lower=0)
pitching_df['service_time'] = (pitching_df['Year'] - pitching_df['debut_year']).clip(lower=0)

# 分類隔年適用的年資類別（pre-arb / arb / FA）
def classify_for_next_year(service_end_current):
    if np.isnan(service_end_current):
        return 'unknown'
    if service_end_current < 3:
        return 'pre-arb'
    elif service_end_current < 6:
        return 'arb'
    else:
        return 'FA'

batting_df['status_next_year'] = batting_df['service_time'].apply(classify_for_next_year)
pitching_df['status_next_year'] = pitching_df['service_time'].apply(classify_for_next_year)

# 做 one-hot encoding
batting_df = pd.get_dummies(
    batting_df,
    columns=['status_next_year'],
    prefix='status',          # 產生的欄位會變成 status_pre-arb, status_arb, status_FA
    drop_first=False          
)

pitching_df = pd.get_dummies(
    pitching_df,
    columns=['status_next_year'],
    prefix='status',          # 產生的欄位會變成 status_pre-arb, status_arb, status_FA
    drop_first=False          
)

print(batting_df.head())
print(pitching_df.head())
print(batting_df[batting_df.name_clean=='mookiebetts'].sort_values('Year')[['name_clean', 'Year', 'Salary', 'next_year_salary', 'WAR', 'L-WAR']])
```

# 特徵工程


```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')

# 打者特徵

# 1. 定義打者「當季核心特徵」
batter_core_features = [
    'Age',                              # 年齡
    'service_time',                     # 當季結束時年資
    'G', 'PA', 'AB',                    # 出賽量
    'R', 'H', 'HR', 'RBI', 'SB',        # 基本產出
    'BB', 'SO',
    'AVG', 'OBP', 'SLG', 'OPS',         # 傳統打擊三圍
    'ISO', 'BABIP',
    'wOBA', 'wRC+', 'WAR', 'L-WAR',     # 進階綜合指標
    'BB%', 'K%', 'BB/K',
    'Barrel%', 'HardHit%', 'EV', 'maxEV', # Statcast 力量
    'xBA', 'xSLG', 'xwOBA'              # Statcast 預期
]

# 加入 status的one-hot 欄位
status_columns = [col for col in batting_df.columns if col.startswith('status_')]
batter_core_features += status_columns

# 只保留實際存在的欄位
batter_features = [f for f in batter_core_features if f in batting_df.columns]
batting_df = batting_df[batting_df['next_year_log_salary'].notna()].copy()

X_batter = batting_df[batter_features].fillna(0)
y_batter = batting_df['next_year_log_salary']

print(f"打者初始特徵數：{len(batter_features)}")

# 2. 使用 RandomForest 計算重要度
rf = RandomForestRegressor(n_estimators=120, random_state=42, n_jobs=-1)
rf.fit(X_batter, y_batter)

importances = pd.Series(rf.feature_importances_, index=batter_features)
importances = importances.sort_values(ascending=False)

print("\n打者當季特徵重要度 Top 25：")
print(importances.head(25))



# 畫熱圖檢查共線性

# 只看數值型特徵（排除 one-hot）
numeric_features = [f for f in batter_features if f not in 
                    [col for col in batting_df.columns if col.startswith(('status_'))]]

corr_matrix = batting_df[numeric_features + ['next_year_log_salary']].corr()

# 畫熱圖
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("特徵與目標的相關性熱圖")
plt.show()

# 手動刪除高共線性特徵（刪了 L-WAR, wOBA, AB, H, BB%, G）
b_final = [
    'service_time', 'wRC+', 'Age', 'status_pre-arb', 'PA', 'WAR',
    'RBI', 'R', 'BABIP', 'SB', 'SO', 'K%', 'BB', 'BB/K',
    'HR', 'xBA', 'maxEV', 'HardHit%', 'SLG'   
]






# 1. 定義投手「當季核心特徵」
pitcher_core_features = [
    'Age',
    'service_time',
    'G', 'GS', 'IP',                    # 出賽量與先發數
    'ERA', 'FIP', 'xFIP', 'SIERA',      # 核心 ERA 相關指標
    'W', 'SV',
    'WHIP', 'K/9', 'BB/9', 'K-BB%',     # 控球與三振
    'HR/9',
    'K%', 'BB%',
    'BABIP', 'LOB%',                    # 運氣與殘壘率
    'ERA-', 'FIP-',                     # 聯盟調整版
    'WAR', 'L-WAR',                     # 綜合貢獻
    'EV', 'HardHit%', 'Barrel%', 'xwOBA'# Statcast
]

# 加入 status的one-hot 欄位
status_columns = [col for col in pitching_df.columns if col.startswith('status_')]
pitcher_core_features += status_columns

# 只保留存在的欄位
pitcher_features = [f for f in pitcher_core_features if f in pitching_df.columns]
pitching_df = pitching_df[pitching_df['next_year_log_salary'].notna()].copy()

X_pitcher = pitching_df[pitcher_features].fillna(0)
y_pitcher = pitching_df['next_year_log_salary']

print(f"投手初始特徵數：{len(pitcher_features)}")

# 2. RandomForest 重要度
rf_p = RandomForestRegressor(n_estimators=120, random_state=42, n_jobs=-1)
rf_p.fit(X_pitcher, y_pitcher)

importances_p = pd.Series(rf_p.feature_importances_, index=pitcher_features)
importances_p = importances_p.sort_values(ascending=False)

print("\n投手當季特徵重要度 Top 25：")
print(importances_p.head(25))


# 畫熱圖檢查共線性

# 只看數值型特徵（排除 one-hot）
p_numeric_features = [f for f in pitcher_features if f not in 
                    [col for col in pitching_df.columns if col.startswith(('status_'))]]

p_corr_matrix = pitching_df[p_numeric_features + ['next_year_log_salary']].corr()

# 畫熱圖
plt.figure(figsize=(18, 16))
sns.heatmap(p_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("特徵與目標的相關性熱圖")
plt.show()

# 手動刪除高共線性特徵（刪了 G, FIP, xFIP, BB%, K%, EV, status_arb, FIP-）
p_final = [
    'service_time', 'status_pre-arb', 'Age', 'SIERA', 'IP', 'WAR','BABIP',
    'BB/9', 'K/9', 'HardHit%','LOB%', 'Barrel%', 'K-BB%', 'WHIP', 'W', 
    'HR/9', 'ERA' 
    
]
```

# 訓練模型

## 打者模型實驗記錄

### v1 - Baseline XGBoost
- 參數：n_estimators=300, lr=0.03, max_depth=5
- 訓練 R²: 0.9137
- 測試 R²: 0.2871
- 問題：嚴重過擬合

### v2 - 加強正則化
- 參數：n_estimators=600, lr=0.008, max_depth=4, reg_lambda=3
- 訓練 R²: 0.7186
- 測試 R²: 0.2933
- 改善：差距從 0.626 → 0.425

### v3 - 繼續加強
- 參數：n_estimators=800, lr=0.005, max_depth=3, reg_lambda=15.0
- 訓練 R²: 0.5858
- 測試 R²: 0.2820
- 改善：過擬合繼續被壓制（差距從 0.425 → 0.304）但測試 R² 不升反降（0.2933 → 0.2820）underfitting 開始出現

### v4 - 參數折衷
- 參數：n_estimators=700, lr=0.007, max_depth=4, reg_lambda=12.0
- 訓練 R²: 0.5858
- 測試 R²: 0.2820
- 改善：過擬合繼續被壓制（差距從 0.425 → 0.304）但測試 R² 不升反降（0.2933 → 0.2820）underfitting 開始出現


```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error


# 共用評估函數
def evaluate_model(y_true, y_pred, title=""):
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{title}")
    print(f"  R²    : {r2:.4f}")
    print(f"  RMSE  : {rmse:.4f}")
    return r2, rmse

# 畫預測 vs 真實值散點圖
def plot_predictions(y_true, y_pred, title=""):
    plt.figure(figsize=(8, 6))
    plt.scatter(y_true, y_pred, alpha=0.6)
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
    plt.xlabel("True next_year_log_salary")
    plt.ylabel("Predicted")
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.show()

# 打者最終特徵（請替換成你實際的清單）
b_final = [
    'service_time', 'wRC+', 'Age', 'status_pre-arb', 'PA', 'WAR',
    'RBI', 'R', 'BABIP', 'SB', 'SO', 'K%', 'BB', 'BB/K',
    'HR', 'xBA', 'maxEV', 'HardHit%', 'SLG'   
]


# 準備資料
X_batter = batting_df[b_final].fillna(0)
y_batter = batting_df['next_year_log_salary']

# 時間序列切分（建議 2021 年以前訓練，之後測試）
train_mask = batting_df['Year'] <= 2021
test_mask  = batting_df['Year'] > 2021

X_train_b = X_batter[train_mask]
y_train_b = y_batter[train_mask]
X_test_b  = X_batter[test_mask]
y_test_b  = y_batter[test_mask]

print(f"打者訓練樣本：{len(X_train_b)}，測試樣本：{len(X_test_b)}")

# batter v1 XGBoost 模型
# xgb_batter = XGBRegressor(
#     n_estimators=300,
#     learning_rate=0.03,
#     max_depth=5,
#     subsample=0.8,
#     colsample_bytree=0.8,
#     random_state=42,
#     n_jobs=-1,
#     objective='reg:squarederror'
# )


# batter v2 加強正則化的版本 
# xgb_batter = XGBRegressor(
#     n_estimators=600,
#     learning_rate=0.008,
#     max_depth=4,
#     min_child_weight=3,
#     gamma=0.05,
#     subsample=0.75,
#     colsample_bytree=0.75,
#     reg_lambda=3.0,          # 加 L2 正則
#     reg_alpha=0.5,
#     random_state=42,
#     n_jobs=-1
# )

# batter v3 繼續加強
# xgb_batter = XGBRegressor(
#     n_estimators=800,
#     learning_rate=0.005,
#     max_depth=3,                # 再降低
#     min_child_weight=10,
#     gamma=0.3,
#     subsample=0.6,
#     colsample_bytree=0.6,
#     reg_lambda=15.0,            # 再提高
#     reg_alpha=2.0,
#     random_state=42,
#     n_jobs=-1
# )

# batter v4 參數折衷
xgb_batter = XGBRegressor(
    n_estimators=700,
    learning_rate=0.007,          # 介於 0.005 和 0.008
    max_depth=4,                  # 回到 4，但不回 5
    min_child_weight=6,           # 介於 5~10
    gamma=0.1,
    subsample=0.7,
    colsample_bytree=0.7,
    reg_lambda=12.0,              # 介於 10 和 15
    reg_alpha=0.5,
    random_state=42,
    n_jobs=-1
)






xgb_batter.fit(
    X_train_b, y_train_b,
    verbose=False
)


# 預測與評估
y_pred_train_b = xgb_batter.predict(X_train_b)
y_pred_test_b  = xgb_batter.predict(X_test_b)

evaluate_model(y_train_b, y_pred_train_b, "打者 - 訓練集表現")
evaluate_model(y_test_b,  y_pred_test_b,  "打者 - 測試集表現")

# 畫圖
plot_predictions(y_test_b, y_pred_test_b, "打者 - 測試集預測 vs 真實 (log salary)")

# 特徵重要度（XGBoost 版本）
from xgboost import plot_importance
plot_importance(xgb_batter, max_num_features=20)
plt.show()
    
```
